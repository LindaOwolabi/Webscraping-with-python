```python
import requests
from bs4 import BeautifulSoup #parse
```

```python
url_link = "https://www.rightmove.co.uk/property-to-rent/find.html?searchType=RENT&locationIdentifier=REGION%5E904&insId=1&radius=0.0&minPrice=&maxPrice=&minBedrooms=&maxBedrooms=&displayPropertyType=&maxDaysSinceAdded=&sortByPriceDescending=&_includeLetAgreed=on&primaryDisplayPropertyType=&secondaryDisplayPropertyType=&oldDisplayPropertyType=&oldPrimaryDisplayPropertyType=&letType=&letFurnishType=&houseFlatShare="
```

```python
r = requests.get(url_link)
print(r.status_code)
```

```python
soup_ = BeautifulSoup(r.content, 'html.parser')
soup_
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
soup
```

```python
# This gets the list of apartments
apartments = soup.find_all("div", class_="l-searchResult is-list")
print(len(apartments))
```

```python
a_info_description = apartments[1].find("span",  {'itemprop': 'description'})
a_info_description = a_info_description.get_text().strip()
a_info_description
```

```python
apt_links = []
apt_addressses = []
apt_info = []
apt_descriptions = []
apt_prices = []
```

```python
for i in range(len(apartments)):

  # tracks which apartment we are on in the page
  apartment_no = apartments[i]

  # append links
  a_info_link = apartment_no.find("a", class_="propertyCard-link")
  a_link = "https://www.rightmove.co.uk" + a_info_link.attrs["href"]
  apt_links.append(a_link)

  # append address
  a_info_address = apartment_no.find("address", class_="propertyCard-address")
  a_info_address = a_info_address.get_text().replace('\n','')
  apt_addressses.append(a_info_address)

  # append info type
  a_info_type = apartment_no.find("h2", class_="propertyCard-title")
  a_info_type = a_info_type.get_text().strip()
  apt_info.append(a_info_type)

  # append description
  a_info_description = apartment_no.find("span",  {'data-test': 'property-description'})
  a_info_description = a_info_description.get_text().strip()
  apt_descriptions.append(a_info_description)

  # append price
  a_info_price = apartment_no.find("span", class_="propertyCard-priceValue")
  a_info_price = a_info_price.get_text().strip()
  apt_prices.append(a_info_price)
```

```python
# convert data to dataframe
data = {
    "Links": apt_links,
    "Address": apt_addressses,
    "Info": apt_info,
    "Description": apt_descriptions,
    "Price": apt_prices,
}
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(data)
```
